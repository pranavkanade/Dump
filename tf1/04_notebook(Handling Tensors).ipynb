```python
import matplotlib.image as mp_image
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
file_cat = "cat.png"
```

```python
cat_image_in = mp_image.imread(file_cat)
```

```python
# dimenstions of the image
print("input dimensions = {}".format(cat_image_in.ndim))
```

```python
# shape of the image
print("input shape = {}".format(cat_image_in.shape))
```

```python
plt.imshow(cat_image_in)
plt.show()
```

```python
cat_image = tf.placeholder("float64", [None, None, 4])
```

```python
slice = tf.slice(cat_image, [10, 0, 0], [16, -1, -1])
```

```python
with tf.Session() as sess:
    result = sess.run(slice, feed_dict={cat_image: cat_image_in})
    print(result.shape)
```

```python
plt.imshow(result)
plt.show()
```

```python
x = tf.Variable(cat_image_in, name="x")
model = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    x = tf.transpose(x, perm=[1, 0, 2])
    sess.run(model)
    result = sess.run(x)
```

```python
plt.imshow(result)
plt.show()
```
